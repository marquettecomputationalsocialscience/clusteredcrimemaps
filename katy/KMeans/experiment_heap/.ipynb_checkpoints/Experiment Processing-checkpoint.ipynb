```python
import os 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.plotly as py
import plotly.figure_factory as ff
%matplotlib inline
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
frame = pd.read_csv('all_data', sep=';', index_col=False)
```

```python
frame.sort_values(['name','k'])
index = frame.columns[2:]
frame['Average'] = frame.loc[:,index].mean(1)
```

```python
frame = frame[frame['Average'] != 0]
frame
```

```python
df = pd.read_json('../bias.json')
index_m = df.columns[1:13]
index_t = df.columns[13:-1]
```

```python
def reorder(a):
    return [a[4],a[3],a[7],a[0],a[8],a[6],a[5],a[1],a[11],a[10],a[9],a[2]]
color_list = ['#66c2a5','#fc8d62','#8da0cb']
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December']
layout = go.Layout(
    autosize=False,
    width=800,
    height=500,
title='Average Potential Bias per Month for Theft' ,
xaxis=dict(
    title='Month'
),
yaxis=dict(
    title='Average Potential Bias',
    range=[0,.14]
),
)
data =[]
for ind,k in enumerate([2,5,10]):
    
    var = df[df.k==k][index_t].mean()
    
    data.append(go.Scatter(y=reorder(var),x=months,
                           mode = 'lines+markers',name = str(k) + ' Clusters',
                          marker={'color': color_list[ind]} ))
fig = go.Figure(data=data,layout=layout)
iplot(fig, filename='Linegraph', image='png')
```
