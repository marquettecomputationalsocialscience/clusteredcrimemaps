```python
import os
import shutil
import calendar
import csv
import pandas
import numpy
```

```python
# pull in griffin's data from his data directory
griffinDataPath = "../griffin/data/"
griffinData = os.listdir(griffinDataPath)
for item in griffinData:
    if os.path.isdir(griffinDataPath + item) and item.startswith("data_"):
        year = item[-4:]
        monthFiles = os.listdir(griffinDataPath + item)
        monthFiles = [m for m in monthFiles if m.endswith(".csv")]
        #print(len(monthFiles))
        for file in monthFiles:
            newFileName = file[:-4] + "_" + year + ".csv"
            shutil.copyfile(griffinDataPath + item + "/" + file, "./raw_data/" + newFileName)
            #print(griffinDataPath + item + "/" + file)
            #print("./clean_data/" + newFileName)
```

```python
rawData = os.listdir("./raw_data/")
#yearlyData = [file for file in rawData if file.startswith("20")]
monthlyData = [file for file in rawData if file[-8:].startswith("20")]
```

```python
# standardize file names
monthAbbr = {"april": 4, 
             "aug": 8, 
             "dec": 12, 
             "feb": 2, 
             "jan": 1, 
             "july":7, 
             "june": 6, 
             "mar": 3, 
             "may": 5, 
             "nov": 11, 
             "oct": 10, 
             "sep": 9}
for fileName in monthlyData:
    nameParts = fileName[:-4]
    nameParts = nameParts.split("_")
    year = nameParts[-1]
    month = monthAbbr[nameParts[-2]]
    if len(nameParts) == 4:
        crime = nameParts[0] + nameParts[1]
    else:
        crime = nameParts[0]
    newName = "%s_%02d_%s.csv" % (year, month, crime)
    os.rename("./raw_data/"+fileName, "./raw_data/"+newName)
```

```python
# bring in justin's data
# note that it has to be converted from comma delimited to semicolon deliminated
justinCrimeTypes = ["ROBBERY", "SIMPLE ASSAULT"]
justinRobberyPath = "../griffin/data/yearly_data/ROBBERY/latlng/"
justinAssaultPath = "../griffin/data/yearly_data/SIMPLE ASSAULT/latlng/"

for crime in justinCrimeTypes:
    justinDataPath = "../griffin/data/yearly_data/"+ crime +"/latlng/"
    yearFiles = os.listdir(justinDataPath)
    for yearFile in yearFiles:
            
        with open(justinDataPath + yearFile, mode="r") as infile:
            reader = csv.reader(infile, delimiter=",")    
            with open("./raw_data/year_" + yearFile[:-8] + crime.lower().replace(" ","") + ".csv", mode="w", newline='') as outfile:
                writer = csv.writer(outfile, delimiter=';')
                for row in reader:
                    Location = [float(row[11]), float(row[12])]
                    newRow = [row[0], 
                              row[1], 
                              row[2], 
                              row[3], 
                              row[4], 
                              row[5], 
                              row[6],
                              row[7],
                              row[8],
                              Location,
                              row[9],
                             ]
                    writer.writerow(newRow)
```

```python
# split up justin's yearly data into monthly files

yearlyData = os.listdir("./raw_data/")
yearlyData = [f for f in yearlyData if f.startswith("year_")]


for yearcrime in yearlyData:
    print(yearcrime)
    year = yearcrime.split("_")[-2]
    crime = yearcrime.split("_")[-1][:-4][-7:]
    
    with open("./raw_data/" + yearcrime, mode="r") as infile:
        reader = csv.reader(infile, delimiter=";")
        
        with open("./raw_data/"+year+"_01_"+crime+".csv", mode="w", newline='') as janfile:
            janwriter = csv.writer(janfile, delimiter = ";")
            with open("./raw_data/"+year+"_02_"+crime+".csv", mode="w", newline='') as febfile:
                febwriter = csv.writer(febfile, delimiter = ";")
                with open("./raw_data/"+year+"_03_"+crime+".csv", mode="w", newline='') as marfile:
                    marwriter = csv.writer(marfile, delimiter = ";")
                    with open("./raw_data/"+year+"_04_"+crime+".csv", mode="w", newline='') as aprfile:
                        aprwriter = csv.writer(aprfile, delimiter = ";")
                        with open("./raw_data/"+year+"_05_"+crime+".csv", mode="w", newline='') as mayfile:
                            maywriter = csv.writer(mayfile, delimiter = ";")
                            with open("./raw_data/"+year+"_06_"+crime+".csv", mode="w", newline='') as junfile:
                                junwriter = csv.writer(junfile, delimiter = ";")
                                with open("./raw_data/"+year+"_07_"+crime+".csv", mode="w", newline='') as julfile:
                                    julwriter = csv.writer(julfile, delimiter = ";")
                                    with open("./raw_data/"+year+"_08_"+crime+".csv", mode="w", newline='') as augfile:
                                        augwriter = csv.writer(augfile, delimiter = ";")
                                        with open("./raw_data/"+year+"_09_"+crime+".csv", mode="w", newline='') as sepfile:
                                            sepwriter = csv.writer(sepfile, delimiter = ";")
                                            with open("./raw_data/"+year+"_10_"+crime+".csv", mode="w", newline='') as octfile:
                                                octwriter = csv.writer(octfile, delimiter = ";")
                                                with open("./raw_data/"+year+"_11_"+crime+".csv", mode="w", newline='') as novfile:
                                                    novwriter = csv.writer(novfile, delimiter = ";")
                                                    with open("./raw_data/"+year+"_12_"+crime+".csv", mode="w", newline='') as decfile:
                                                        decwriter = csv.writer(decfile, delimiter = ";")
                                                        
                                                        for row in reader:
                                                            month = int(row[1].split("/")[0])
                                                            
                                                            if month == 1:
                                                                janwriter.writerow(row)
                                                            elif month == 2:
                                                                febwriter.writerow(row)
                                                            elif month == 3:
                                                                marwriter.writerow(row)
                                                            elif month == 4:
                                                                aprwriter.writerow(row)
                                                            elif month == 5:
                                                                maywriter.writerow(row)
                                                            elif month == 6:
                                                                junwriter.writerow(row)
                                                            elif month == 7:
                                                                julwriter.writerow(row)
                                                            elif month == 8:
                                                                augwriter.writerow(row)
                                                            elif month == 9:
                                                                sepwriter.writerow(row)
                                                            elif month == 10:
                                                                octwriter.writerow(row)
                                                            elif month == 11:
                                                                novwriter.writerow(row)
                                                            elif month == 12:
                                                                decwriter.writerow(row)
    os.remove("./raw_data/" + yearcrime)
                                                            
                                                        
```
