```python
# copy and rename justin's data (doesn't need to be run again)
robbery_assult_dir = '../../griffin/KMeans/justin_datamound/'
robbery_assult_files = os.listdir('../../griffin/KMeans/justin_datamound/')
for filename in robbery_assult_files:
    [year, monthNum, crimeType] = filename.split("_")
    crimeType = crimeType.replace('.js', '')
    crimeType = crimeType.replace(' ', '')
    crimeType = crimeType.lower()
    monthAbbr = calendar.month_abbr[int(monthNum)].lower()
    shutil.copyfile(robbery_assult_dir + filename, './datamound/data_'+year+'-'+crimeType+'_'+monthAbbr+'.js')
    print('data_'+year+'-'+crimeType+'_'+monthAbbr+'.js')
    
```
